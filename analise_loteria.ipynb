```python
import pandas as pd
import random
```

```python
df = pd.read_excel("C:/Users/Usuário/Desktop/lotofacil.xlsx")

lista_dezenas = []
lista_repeticao = []
lista_dezenas_menos = []
lista_dezenas_mais = []

for i in range(1,26):
    valores = df[(df[['bola 1','bola 2','bola 3','bola 4','bola 5', 'bola 6','bola 7','bola 8','bola 9','bola 10','bola 11','bola 12','bola 13','bola 14','bola 15']] == i)].count()
    lista_repeticao.append(valores.sum())
    lista_dezenas.append(i)

dados = {"Dezena":lista_dezenas,"Repetição":lista_repeticao}

df_filtros = pd.DataFrame(dados)


display(df_filtros.sort_values(by="Repetição"))

for i, repeticao in enumerate(df_filtros['Repetição']):
    dezenas = int(df_filtros.loc[i,'Dezena'])
    if repeticao >= 1760:
        lista_dezenas_mais.append(dezenas)
    elif repeticao < 1760:
        lista_dezenas_menos.append(dezenas)
```

```python
lista_jogos = []
for i in range(0,6):
    dezenas_mais = random.sample(lista_dezenas_mais,11)
    dezenas_menos = random.sample(lista_dezenas_menos,4)
    
    lista_jogos.extend(dezenas_mais)
    lista_jogos.extend(dezenas_menos)
    
    print(sorted(lista_jogos))
    lista_jogos.clear()
```
